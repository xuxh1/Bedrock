```python
import os
import numpy as np
import xarray as xr
import pandas as pd
import netCDF4 as nc
import geopandas as gpd
import rioxarray as rxr
from myfunc import timer
from myfunc import DirMan
from myfunc import load_and_flatten_data
import config

# configuration
# resolution = "0p1"
resolution = "500"
region = [-180,180,-60,90]
# region = [-124.8,-66.95,24.5,49.4]
data_path = f'/tera04/zhwei/xionghui/bedrock/run/{resolution}/'
post_data_path = '/tera04/zhwei/xionghui/bedrock/'
shp_path = '/tera04/zhwei/xionghui/bedrock/Shp/'
fig_path = f'/home/xuxh22/stu01/Bedrock/fig/{resolution}/'
path = '/home/xuxh22/stu01/Bedrock/'
if resolution == "0p1":
    size = 0.1
elif resolution == "500":
    size = 0.0005

dir_man = DirMan(data_path)
dir_man.enter()

os.makedirs(f'{data_path}/csv', exist_ok=True)

casename1 = 'bedrock_1'
path1 = f'{data_path}/cases/{casename1}/'
casename2 = 'bedrock_2'
path2 = f'{data_path}/cases/{casename2}/'
casename3 = 'bedrock_3'
path3 = f'{data_path}/cases/{casename3}/'
```

```python
df = pd.DataFrame()

with nc.Dataset(f'Sbedrock.nc4') as dataset:
    lat = dataset['lat'][:].flatten()
    lon = dataset['lon'][:].flatten()
    df['lat'] = np.repeat(lat, len(lon))
    df['lon'] = np.tile(lon, len(lat))

file_variable_list = [
    (f'mask1234', 'Band1'),
    # (f'Sbedrock', 'Sbedrock'),
    (f'Sbedrock', 'Band1'),
    (f'Area', 'area'),

    (f'{path1}Sbedrock_mean', 'Sr'),
    (f'{path2}Sbedrock_mean', 'Sr'),
    (f'{path3}Sbedrock_mean', 'Sr'),
]

for entry in file_variable_list:
    file = entry[0]
    variable_name = entry[1]  
    index = entry[2:] if len(entry) > 2 else None  

    # print(f"File: {file}, Variable: {variable_name}, Index: {index}")
    if index:
        df[file] = load_and_flatten_data(file, variable_name, index[0])
    else:
        df[file] = load_and_flatten_data(file, variable_name)


# df = df.dropna()
# df = df[df['Sbedrock'] > 0]
df = df[df['mask1234'] == 1]
df.drop(labels='mask1234',axis=1,inplace=True)

print(df[f'{path1}Sbedrock_mean'].mean())
print(df[f'{path2}Sbedrock_mean'].mean())
print(df[f'{path3}Sbedrock_mean'].mean())
print(df['Sbedrock'].mean())

with open(f'csv/Global_cases_sb.csv','w') as f:
    df.to_csv(f)
df
```
